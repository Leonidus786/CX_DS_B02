```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
html = response.text

tables = pd.read_html(html, match="Country", header=0)

df = tables[0]
print(df.head())
```

```python
def fetch_tables(url, match=None):
    import requests
    import pandas as pd

    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        raise Exception(f"Failed to fetch page: {response.status_code}")

    return pd.read_html(response.text, match=match)
```

```python
df = fetch_tables(url, match="Country")[0]
df.head()
```
