#### Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

## The main difference between .loc and .iloc in pandas is that .loc is label-based (using row/column names), while .iloc is integer position-based (using 0-based numerical indices). .loc includes the stop value in slices, whereas .iloc excludes it, behaving like standard Python lists. 

### Key Differences:
## .loc` (Label-based):
- Uses index labels and column names.
- Includes the last item in a slice [start:stop].

## Can use boolean masks/arrays.
- Raises KeyError if labels are not found.
- Example: df.loc[0:2, 'Column1'] (includes index 2).

## .iloc` (Integer-positional):
- Uses integer positions (0 to length-1).
- Excludes the last item in a slice [start:stop].
- Cannot use boolean masks.
- Raises IndexError if indices are out of bounds.
- Example: df.iloc[0:2, 1] (includes index 0 and 1, stops before 2). 

### When to Use Which:
- Use .loc when you know the specific names of rows or columns, or when filtering based on conditions (e.g., df.loc[df['A'] > 5]).
- Use .iloc when you know the numerical position of rows or columns, or need to slice based on position.
